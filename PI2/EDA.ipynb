#### Analisis de la Penentracion de internet
    En este analisis me centrare en las hojas relacionadas a la penetracion de internet en la poblacion y los hogares argentinos 


```python
# Carga
P_hogares=pd.read_csv('C:/Users/Caeolina/Downloads/PROYECTO_INDIVIDUAL2-master/PI2/Penetracion-hogares_procesado.csv', delimiter=',')
p_poblacion = pd.read_csv('C:/Users/Caeolina/Downloads/PROYECTO_INDIVIDUAL2-master/PI2/Penetracion-poblacion_procesado.csv', delimiter=',')
```

##### GRAFICO 1: MAPA DE CALOR (HEATMAP)

```python
# Mapa de calor de penetracin por provincia

pivot = p_poblacion.pivot_table(index='Provincia', columns='Ao', values='Accesos por cada 100 hab', aggfunc='mean')
sns.heatmap(pivot, annot=True, cmap='coolwarm', fmt='.1f')
plt.title('Penetracin de Internet en la Poblacion por Provincia')
plt.show()
```

##### Proposito:
    el propocito de este mapa es visualizar la penetracion de internet en la poblacion por provincia, para identificar provincias con baja penetracion en donde seria conveniente priorisar las inverciones de servicio de internet debido a la demanda en la sona.
    Como se puede apreciar en el mapa las probincias con mas baja penetracion en el ao 2024 fueron Formosa, Chaco, Santiago del Estero.
    Tambien podemos concluir que Capital federal no seria un buen lugar para priorizar inverciones debido a que su penetracion es la mas alta, con mucha diferencia.

##### GRAFICO 2: GRAFICO DE BARRAS:
    

```python
P_hogares.groupby('Provincia')['Accesos por cada 100 hogares'].mean().sort_values().plot(kind='barh', figsize=(10, 6), color='skyblue')
plt.title('Penetracin de Internet en Hogares por Provincia')
plt.xlabel('Porcentaje de Penetracin')
plt.ylabel('Provincia')
plt.show()
```

##### Propocito:
    El propocito de este grafico es comparar la penetracion de internet en hogares entre provincias, con el fin de identificar provincias con menor penetracion de servicio de internet en hogares.
    Como podemos ver en este grafico, las provincias de Formosa, Chaco y Santiago del estero son las provincias con mas baja penetracion de servicios de internet en hogares con menos de un 40%. 
    
    Si comparamos la penetracion en la poblacion y la penetracion en los hogares del servicio por provincia, podemos observar que las provincias con mas baja penetracion coinciden en ambos graficos, lo cual seria un indicativo de que estas podrian encabezar la lista de prioridad para que la empresa invierta en mejorar los servicios de internet gracias a la demanda que se genera.

##### GRAFICO 2: GRAFICO DE LINEAS

    Realiz este grafico con el fin de identificar tendencias de mejora o estancamiento en la velocidad de conexion durante los aos. Como podemos ver, no hubo un estancamiento en si, pero si una fluctuacion importante en el 2021 con la caida rapida de la velocidad media descargada cuando venia en aumento de forma rapida en los aos anteriores y luego de este suceso volvi a repuntar pero de manera mas constante. 

#### Analisis de accesos por tecnologia:

##### GRAFICO 1: GRAFICO DE BARRAS APILADAS

```python
Accesos_tec=pd.read_csv('C:/Users/Caeolina/Downloads/PROYECTO_INDIVIDUAL2-master/PI2/Accesos Por Tecnologa_procesado.csv', delimiter=',')
```

```python
# Agrupar por Provincia y sumar los valores de las tecnologas
df_agrupado = Accesos_tec.groupby('Provincia')[['ADSL', 'Cablemodem', 'Fibra ptica', 'Wireless', 'Otros']].sum()
# Configuramos el grfico de barras apiladas
df_agrupado.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Accesos por Tecnologa y Provincia')
plt.xlabel('Provincia')
plt.ylabel('Nmero de Accesos')
plt.show()
```

#### Analisis de Ingresos
    En este analisis nos centraremos en las hojas relacionadas a los ingresos generados por los servicios de internet.

```python
Ingresos=pd.read_csv('C:/Users/Caeolina/Downloads/PROYECTO_INDIVIDUAL2-master/PI2/Ingresos_procesado.csv', delimiter=',')
```

##### GRAFICO 1: GRAFICO DE LINEAS 

```python
# Evolucin de los ingresos
Ingresos.groupby('Ao')['Ingresos (miles de pesos)'].sum().plot(kind='line', figsize=(10, 6), color='blue')
plt.title('Evolucin de los Ingresos por Servicios de Internet')
plt.xlabel('Ao')
plt.ylabel('Ingresos (miles de pesos)')
plt.show()
```

##### Propocito: 
    Mostrar la evolucion de los ingresos a lo largo del tiempo, Para reconocer tendencias de crecimiento o declive de los ingresos. En este grafico se ve claramente que hubo un crecimiento progrecibo a lo largo de los aos y a partir de 2021 tubo un acenso mas pronunciado hasta 2024, donde se produjo un declive de los ingresos. 

##### GRAFICO 2: GRAFICO DE BARRAS

```python
# Ingresos por Trimestre
Ingresos.groupby(' Trimestre ')['Ingresos (miles de pesos)'].sum().sort_values().plot(kind='barh', figsize=(10, 6), color='orange')
plt.title('Ingresos por  Trimestre ')
plt.xlabel('Ingresos (miles de pesos)')
plt.ylabel(' Trimestre ')
plt.show()
```

##### Propocito:
    En este grafico podemos comparar los ingresos por trimestre a lo largo de los aos, se ve claramente que el trimestre con mas contribucion a lo largo de los aos es el segudo en los servicios de internet, por lo cual es util a la hora de lanzar el servicio al mercado saber cual es el mejor momento del ao, donde el consumidor esta mas dispuesto a adquirir un servicio.
