#### Analisis de la Penentracion de internet
    En este analisis me centrare en las hojas relacionadas a la penetracion de internet en la poblacion y los hogares argentinos 


```python
# Carga
P_hogares=pd.read_csv('C:/Users/Caeolina/Downloads/PROYECTO_INDIVIDUAL2-master/PI2/Penetracion-hogares_procesado.csv', delimiter=',')
p_poblacion = pd.read_csv('C:/Users/Caeolina/Downloads/PROYECTO_INDIVIDUAL2-master/PI2/Penetracion-poblacion_procesado.csv', delimiter=',')
```

##### GRAFICO 1: MAPA DE CALOR (HEATMAP)

```python
# Mapa de calor de penetración por provincia

pivot = p_poblacion.pivot_table(index='Provincia', columns='Año', values='Accesos por cada 100 hab', aggfunc='mean')
sns.heatmap(pivot, annot=True, cmap='coolwarm', fmt='.1f')
plt.title('Penetración de Internet en la Poblacion por Provincia')
plt.show()
```

##### Proposito:
    el propocito de este mapa es visualizar la penetracion de internet en la poblacion por provincia, para identificar provincias con baja penetracion en donde seria conveniente priorisar las inverciones de servicio de internet debido a la demanda en la sona.
    Como se puede apreciar en el mapa las probincias con mas baja penetracion en el año 2024 fueron Formosa, Chaco, Santiago del Estero.
    Tambien podemos concluir que Capital federal no seria un buen lugar para priorizar inverciones debido a que su penetracion es la mas alta, con mucha diferencia.

##### GRAFICO 2: GRAFICO DE BARRAS:
    

```python
P_hogares.groupby('Provincia')['Accesos por cada 100 hogares'].mean().sort_values().plot(kind='barh', figsize=(10, 6), color='skyblue')
plt.title('Penetración de Internet en Hogares por Provincia')
plt.xlabel('Porcentaje de Penetración')
plt.ylabel('Provincia')
plt.show()
```

##### Propocito:
    El propocito de este grafico es comparar la penetracion de internet en hogares entre provincias, con el fin de identificar provincias con menor penetracion de servicio de internet en hogares.
    Como podemos ver en este grafico, las provincias de Formosa, Chaco y Santiago del estero son las provincias con mas baja penetracion de servicios de internet en hogares con menos de un 40%. 
    
    Si comparamos la penetracion en la poblacion y la penetracion en los hogares del servicio por provincia, podemos observar que las provincias con mas baja penetracion coinciden en ambos graficos, lo cual seria un indicativo de que estas podrian encabezar la lista de prioridad para que la empresa invierta en mejorar los servicios de internet gracias a la demanda que se genera. 

#### Analisis de velocidad de conexion

##### GRAFICO 2: GRAFICO DE LINEAS
    

```python
Vel_sin_rangos=pd.read_csv('C:/Users/Caeolina/Downloads/PROYECTO_INDIVIDUAL2-master/PI2/Velocidad_sin_Rangos_procesado.csv', delimiter=',')
```

```python
# Evolución de la velocidad media descargada
Vel_sin_rangos.groupby('Año')[' Velocidad '].mean().plot(kind='line', figsize=(10, 6), color='green')
plt.title('Evolución de la Velocidad Media Descargada')
plt.xlabel('Año')
plt.ylabel('Velocidad Media (Mbps)')
plt.show()
```

##### Propocito:
    El propocito de este grafico es mostrar la evolucion de la velocidad media descargada (VMD) a lo largo de los años.
    como podemos observar a partir de 2018 hubo un aumento constante hasta 2021 que llego al pico de Mbps descargada y decayo durante ese año hasta 2022 donde repunto nuevamente y aumento de manera mas progresiva hacia el 2024. 

    Realizé este grafico con el fin de identificar tendencias de mejora o estancamiento en la velocidad de conexion durante los años. Como podemos ver, no hubo un estancamiento en si, pero si una fluctuacion importante en el 2021 con la caida rapida de la velocidad media descargada cuando venia en aumento de forma rapida en los años anteriores y luego de este suceso volvió a repuntar pero de manera mas constante. 

#### Analisis de accesos por tecnologia:
    Este analisis se centra en las hojas relacionadas con los accesos a internet por tecnología.

##### GRAFICO 1: GRAFICO DE BARRAS APILADAS

```python
Accesos_tec=pd.read_csv('C:/Users/Caeolina/Downloads/PROYECTO_INDIVIDUAL2-master/PI2/Accesos Por Tecnología_procesado.csv', delimiter=',')
```

```python
# Agrupar por Provincia y sumar los valores de las tecnologías
df_agrupado = Accesos_tec.groupby('Provincia')[['ADSL', 'Cablemodem', 'Fibra óptica', 'Wireless', 'Otros']].sum()
# Configuramos el gráfico de barras apiladas
df_agrupado.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Accesos por Tecnología y Provincia')
plt.xlabel('Provincia')
plt.ylabel('Número de Accesos')
plt.show()
```

    Utilizamos este grafico para comparar accesos a internet por tecnologias en diferentes provincias, para identificar provincias con baja adopcion de tecnologias abanzadas como la fibra optica, la cual es realmente la conexión a Internet más rápida, más fiable y de más alta tecnología que existe. Por lo tanto seria una oportunidad de negocio en invertir en fibra optica en las probincias con menor adopcion de esta tecnologia.

#### Analisis de Ingresos
    En este analisis nos centraremos en las hojas relacionadas a los ingresos generados por los servicios de internet.

```python
Ingresos=pd.read_csv('C:/Users/Caeolina/Downloads/PROYECTO_INDIVIDUAL2-master/PI2/Ingresos_procesado.csv', delimiter=',')
```

##### GRAFICO 1: GRAFICO DE LINEAS 

```python
# Evolución de los ingresos
Ingresos.groupby('Año')['Ingresos (miles de pesos)'].sum().plot(kind='line', figsize=(10, 6), color='blue')
plt.title('Evolución de los Ingresos por Servicios de Internet')
plt.xlabel('Año')
plt.ylabel('Ingresos (miles de pesos)')
plt.show()
```

##### Propocito: 
    Mostrar la evolucion de los ingresos a lo largo del tiempo, Para reconocer tendencias de crecimiento o declive de los ingresos. En este grafico se ve claramente que hubo un crecimiento progrecibo a lo largo de los años y a partir de 2021 tubo un acenso mas pronunciado hasta 2024, donde se produjo un declive de los ingresos. 

##### GRAFICO 2: GRAFICO DE BARRAS

```python
# Ingresos por Trimestre
Ingresos.groupby(' Trimestre ')['Ingresos (miles de pesos)'].sum().sort_values().plot(kind='barh', figsize=(10, 6), color='orange')
plt.title('Ingresos por  Trimestre ')
plt.xlabel('Ingresos (miles de pesos)')
plt.ylabel(' Trimestre ')
plt.show()
```

##### Propocito:
    En este grafico podemos comparar los ingresos por trimestre a lo largo de los años, se ve claramente que el trimestre con mas contribucion a lo largo de los años es el segudo en los servicios de internet, por lo cual es util a la hora de lanzar el servicio al mercado saber cual es el mejor momento del año, donde el consumidor esta mas dispuesto a adquirir un servicio.
